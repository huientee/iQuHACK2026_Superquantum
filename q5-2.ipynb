We begin with the identity
$$
\mathrm{SWAP} = \frac{1}{2}(XX + YY + ZZ + II).
$$

Rearranging gives
$$
H_2 = XX + YY + ZZ = 2\,\mathrm{SWAP} - I.
$$

Substituting this decomposition into the target unitary
$$
U = e^{i\frac{\pi}{4} H_2},
$$
we obtain
$$
U = \exp\!\left(i\frac{\pi}{4}(2\,\mathrm{SWAP} - I)\right).
$$

Since the identity operator commutes with all operators, the exponential factorizes as
$$
U
= \exp\!\left(-i\frac{\pi}{4}\right)
  \exp\!\left(i\frac{\pi}{2}\,\mathrm{SWAP}\right).
$$

The SWAP operator satisfies the involutory property
$$
\mathrm{SWAP}^2 = I.
$$

For any Hermitian operator $M$ such that $M^2 = I$, we have
$$
\exp(i\theta M) = \cos(\theta)\,I + i\sin(\theta)\,M.
$$

Applying this identity with $M = \mathrm{SWAP}$ and $\theta = \frac{\pi}{2}$ yields
$$
\exp\!\left(i\frac{\pi}{2}\,\mathrm{SWAP}\right)
= \cos\!\left(\frac{\pi}{2}\right) I
  + i\sin\!\left(\frac{\pi}{2}\right)\mathrm{SWAP}
= i\,\mathrm{SWAP}.
$$

Combining the results, the target unitary simplifies to
$$
U
= e^{-i\pi/4}\, i\,\mathrm{SWAP}
= e^{i\pi/4}\,\mathrm{SWAP}.
$$

Therefore,
$$
\boxed{
e^{i\frac{\pi}{4} H_2}
= e^{i\pi/4}\,\mathrm{SWAP}
}
$$

Hence, the target unitary is equal to the SWAP gate up to a global phase.
