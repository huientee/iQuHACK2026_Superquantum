# Question 5

We start from the identity

$$
\mathrm{SWAP} = \frac{1}{2}(XX + YY + ZZ + II).
$$

Therefore,

$$
H_2 = XX + YY + ZZ = 2\,\mathrm{SWAP} - I.
$$

Substituting this decomposition into $U$:

$$
U = \exp\!\left(i\frac{\pi}{2}(2\,\mathrm{SWAP} - I)\right).
$$

Since the identity operator commutes with all operators, the exponential can be factorized as

$$
U = \exp\!\left(-i\frac{\pi}{2}\right)
    \exp\!\left(i\pi\,\mathrm{SWAP}\right).
$$

The SWAP operator satisfies the involutory property

$$
\mathrm{SWAP}^2 = I.
$$

For any Hermitian operator $M$ such that $M^2 = I$,

$$
\exp(i\theta M) = \cos(\theta)\,I + i\sin(\theta)\,M.
$$

Applying this identity with $M = \mathrm{SWAP}$ and $\theta = \pi$, we obtain

$$
\exp(i\pi\,\mathrm{SWAP})
= \cos(\pi)\,I + i\sin(\pi)\,\mathrm{SWAP}
= -I.
$$

Combining the results above, the target unitary simplifies to

$$
U = e^{-i\pi/2}(-I) = e^{i\pi/2} I.
$$

Therefore, $U$ equals to the identity operator up to global phase. The optimal Clifford+T implementation is the empty circuit, requiring no quantum gates.
